## Extract relevant city and global data from SQL database

The following SQL statement was used to retrieve global and San Francisco average temperatures. I noticed that recording of global average temperature started in 1750, whereas recording of SF data started about 100 years later in 1849. Therefore, I used an inner join on year to only retrieve the entries for which both SF and global data existed. 

`
SELECT c.year, c.city, c.country, c.avg_temp AS sf_avg_temp, g.avg_temp AS global_avg_temp
FROM city_data AS c
JOIN global_data AS g
ON c.year = g.year
WHERE city = 'San Francisco'
`

The result of the above SQL query was saved as 'global_sf_data.csv'. 

## Import data into python and calculate moving averages

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab

from scipy.stats import pearsonr
```

```python
df = pd.read_csv('global_sf_data.csv')
```

```python
# a function that calculates the moving average
# window of the moving average is passed into the function
def moving_avg(window=10, series=df['sf_avg_temp']):
    return [(series[(i-window):i]).mean() for i in range(len(df))]
```

```python
# select the window for claculating moving average
window = 30
```

```python
# calculate moving average for San Francisco, global and add to dataframe
df['sf_ma'] = moving_avg(window, df['sf_avg_temp'])
df['global_ma'] = moving_avg(window, df['global_avg_temp'])

# calculate ratio of SF and global moving average temperature and add to dataframe
df['ratio_ma'] = df['sf_ma'] / df['global_ma']
```

## Create line plots to visually explore data

```python
def plot_temp(data=df['sf_ma'], line_color='#000000', plot_title='SF average temperature'):
    plt.plot(df['year'], data, color=line_color)
    plt.title(plot_title)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.xlabel('year', fontsize=13)
    plt.ylabel(f"{window}-year moving average", fontsize=13)
    plt.xticks(df['year'][list(range(1,len(df),20))], rotation=90, fontsize=10)
    plt.yticks(fontsize=10)
```

```python
params = {
    'legend.fontsize': 'x-large',
    'figure.figsize': (15, 5),
    'axes.labelsize': 'x-large',
    'axes.titlesize':'x-large',
    'xtick.labelsize':'x-large',
    'ytick.labelsize':'x-large'}

pylab.rcParams.update(params)
```

```python
plt.subplot(1, 3, 1)
plot_temp(df['global_ma'], line_color='#ff0000', plot_title='global average temperature')
plt.subplot(1, 3, 2)
plot_temp(df['sf_ma'], line_color='#000000', plot_title='SF average temperature')
plt.subplot(1, 3, 3)
plot_temp(df['ratio_ma'], line_color='#33cc33', plot_title='ratio of SF and global avg temp')

plt.show()
```

## Summarize observations

The global average temperature has been rising during the last 100 years. The 30-year moving average of the global average temperature has increased by 1.1 °C from 8.15 °C in 1913 to 9.25 °C in 2013. 

---

Average temperatures in San Francisco (SF) have always been higher than global average temperatures. On average the 30-year moving average was about 5.8 °C higher in SF than global. 

---

The difference between SF and global average temperatures has decreased over the last 100 years. In 1913 it was about 1.74 times warmer in SF than global, whereas in 2013 it was only about 1.62 times warmer in SF. This suggests that the global average temperature has increased more than the average temperature in SF over time. 

---

Global average temperature and SF average temperature are highly correlated. The Pearson's correlation coefficient for the 30-year moving averages of global and SF average temperatures during the last 100 years is 0.97.
